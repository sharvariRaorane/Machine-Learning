Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
```

Import dataset

```python
from google.colab import files
uploded = files.upload()
```

```python
import io
data = pd.read_csv(io.BytesIO(uploded['test linear regression.csv']))
```

```python
data
```

```python
X = data['x']
Y = data['y']
```

```python
X = X.values.reshape(len(X),1)
```

```python
Y=Y.values.reshape(len(Y),1)
```

```python
plt.scatter(X, Y) 
plt.title( 'Test Data') 
plt.xlabel('x') 
plt.ylabel('y') 
plt.xticks(()) 
plt.yticks(()) 
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X,Y)
```

```python
plt.plot(X, regr.predict(X), color='black', linewidth = 3)
plt.show()
```
